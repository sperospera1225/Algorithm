```python
class Node(object):
    def __init__(self, val):
        self.val = val
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def get(self, index) -> int:
        if index < 0 or index > self.size:
            return -1
        curr = self.head
        for i in range(index)
             curr = curr.next
        return curr.val
    
    def addAtHead(self, val) -> None:
        node = Node(val)
        node.next = self.head
        self.head = node
        self.size += 1
    
    def addAtTail(self, val) -> None:
        node = Node(val)
        if self.head is None
            self.head = node
        curr = self.head
        while (curr.next != None):
            curr = curr.next
        curr.next = node
        self.size += 1
        
    def addAtIndex(self, index, val) -> None:
        if index < 0 or index > self.size:
            return -1
        if index == 0:
            self.addAtHead(val)
        else:
            node = Node(val)
            curr = self.head
            for i in range(index-1):
                curr = curr.next
            node.next = curr.next
            curr.next = node
            self.size += 1
            
    def deleteAtIndex(self, index) -> None:
        if index < 0 or index >= self.size:
            return
        curr = self.head
        if index == 0:
            self.head = curr.next
        else:
            for i in range(index - 1):
                curr = curr.next
            curr.next = curr.next.next
        self.size -= 1
    
    def deleteNode(self, node):
        curr = self.head
        while curr.next is not None:
            if curr.next == node:
                curr.next=curr.next.next
        self.size -= 1
    
    def find_k(self, k):
        curr = self.head
        for i in range(self.size-k):
            curr = curr.next
        return curr.val
    
            
```
